```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_excel("GE6723_tradebook_2019-07-31_to_2020-06-30.xlsx", sheet_name='TRADEBOOK')
```

```python
order_df = data.iloc[10:,1:11]
order_df.columns = data.iloc[8,1:11]
order_df = order_df[["Trade Date","Symbol","Exchange","Segment","Trade Type","Qty","Price","Time"]]
```

```python
order_data = {}
trade_data = {}
order_df.head(5)
```

```python
trade_types = {
    "Added Long":0,
    "Added Short":1,
    "Partial Selling":2,
    "Square Off Selling":3,
    "Square Off Buying":4,
    "Partial Buying":5,
    "Square Off Buying":6,
    "Over Selling":7,
    "Over Buying":8,
}
trade_duration = {
    "Intraday":0,
    "BTST":1,
    "Short Term":2,
    "Intraday":3,
}
```

```python
order_df.sort_values(["Symbol","Time"], inplace = True)
```

```python
for index, row in order_df.iterrows():
    row = row.tolist()
    symbol = row.pop(1)
    order_data_for_symbol = order_data.get(symbol,[])
    current_order = {
        "Trade Date":row[0],
        "Exchange":row[1],
        "Segment":row[2],
        "Trade Type":row[3],
        "Qty":row[4],
        "Price":row[5],
        "Time":row[6],
        "Order Type":[],
        "Trade Duration":[]
    }
    order_data_for_symbol.append(current_order)
    order_data[symbol]["trades"] = order_data_for_symbol
    holding = order_data[symbol].get("holding")
    if holding and holding.get("Exchange") !=  row.get("Exchange") and holding.get("Segment") !=  row.get("Segment"):
        order_data[symbol]["holding"] = current_order
    else:
        total_quantities = holding.get("Qty") + row.get("Qty")
        if holding.get("Trade Type") == row.get("Trade Type"):
            average_price = (holding.get("Qty")*holding.get("Price") + row.get("Qty")*row.get("Price"))/row.get("Price")+holding.get("Price")
            if holding.get("Trade Type") == "BUY":
                order_type = trade_types["Added Long"]
            else:
                order_type = trade_types["Added Short"]
        else:
            average_price = (holding.get("Qty")*holding.get("Price") - row.get("Qty")*row.get("Price"))/row.get("Price")+holding.get("Price")
            if total_quantities == 0:
                if holding.get("Trade Type") == "SELL":
                    order_type = trade_types["Square Off Buying"]
                else:
                    order_type = trade_types["Square Off Selling"]
            elif total_quantities > 0:
                order_type = trade_types["Over Selling"]
                
            else:
                order_type = trade_types["Over Buying"]
                
       
        
        
```
